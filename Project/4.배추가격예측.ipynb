```r
#데이터 정규화
Pnorm <- (data_ann$평균가격- min(data_ann$평균가격))/(max(data_ann$평균가격)-min(data_ann$평균가격))*0.9 + 0.05
data_ann <- cbind(data_ann, 가격norm=Pnorm)
data_ann
```

```r
test80 <- round(n*0.8, 0)
data_ann_learning <- data_ann[1:test80,]
data_ann_learning
```

```r
head(data_ann_learning)
tail(data_ann_learning)
```

```r
data_ann_test <- data_ann[(test80 + 1):n,]
head(data_ann_test)
tail(data_ann_test)
```

```r
str(data_ann_test)
```

```r
INPUT_NODES <- 10
HIDDEN_NODES <- 5
OUTPUT_NODE <- 5
ITERATION <- 95
```

```r
getDataSet <- function(item,from,to,size){
    dataframe <- NULL
    to <- to - size + 1
    for(i in from:to){
        start <- i
        end <- start + size - 1
        temp <- item[c(start:end)]
        dataframe <- rbind(dataframe, t(temp))
    }
    return(dataframe)
}
```

```r
in_learning <- getDataSet(data_ann_learning$가격norm, 1,996, INPUT_NODES)
in_learning
```

```r
out_learning <- getDataSet(data_ann_learning$가격norm,93, 1083, OUTPUT_NODE)
out_learning
```

```r
#행을 맞춰야 함
model <- nnet(in_learning, out_learning, size = HIDDEN_NODES, maxit = ITERATION)
```

```r
# 데이터프레임 df에서 누락된 값 확인
any(is.na(data_ann))
```

```r
in_test <- getDataSet(data_ann_test$가격norm, 1, 221, INPUT_NODES)
in_test
```

```r
predicted_values <- predict(model, in_test, type="raw")
Vpredicted <- (predicted_values - 0.05)/ 0.9 * (max(data_ann$평균가격) - min(data_ann$평균가격))
+ min(data_ann$평균가격)

Vpredicted

```

```r
Vreal <- getDataSet(data_ann$평균가격,55, 270, OUTPUT_NODE)
Vreal
```

```r
ERROR <- abs(Vreal - Vpredicted)
MAPE <- rowMeans(ERROR/Vreal) * 100
MAPE
```

```r
mean(MAPE)
```

```r
in_forecasting <- data_ann$가격norm[1244:1354]
in_forecasting
```

```r
predicted_values <- predict(model, in_forecasting, type="raw")
Vpredicted <- (predicted_values - 0.05) / 0.9*(max(data_ann$평균가격)-min(data_ann$평균가격))+min(data_ann$평균가격)

Vpredicted
```

```r
# 예측 오차 MAPE #

# 예측값과 실제값이 저장된 벡터
predicted <- c(8933, 9465)
actual <- c(7956, 9452)

# MAPE 계산 함수 정의
calculate_mape <- function(predicted, actual) {
n <- length(predicted)
sum_of_errors <- sum(abs((actual - predicted) / actual))
  mape <- (1 / n) * sum_of_errors * 100
return(mape)
}

# MAPE 계산
mape <- calculate_mape(predicted, actual)
print(paste("MAPE:", mape))

# MSE 계산
mse <- mean((actual - predicted)^2)
print(paste("MSE:", mse))

# MAE 계산
mae <- mean(abs(actual - predicted))
print(paste("MAE:", mae))

# RMSE 계산
rmse <- sqrt(mean((actual - predicted)^2))
print(paste("RMSE:", rmse))

# 결정 계수 계산
r_squared <- 1 - sum((actual - predicted)^2) / sum((actual - mean(actual))^2)
print(paste("R-squared:", r_squared))
```

```r

# 2019~2023 데이터 사용
data_ann_M <- dataResultMean_arima_remove2024 
```

```r
n <- nrow(data_ann_M)
rownames(data_ann_M) <- 1:n
```

```r
#데이터 정규화
Pnorm_M <- (data_ann_M$평균가격- min(data_ann_M$평균가격))/(max(data_ann_M$평균가격)-min(data_ann_M$평균가격))*0.9 + 0.05
data_ann_M <- cbind(data_ann_M, 가격norm=Pnorm_M)
data_ann_M
```

```r
test80_M <- round( n*0.8, 0)
data_ann_learning_M <- data_ann_M[1:test80_M,]
data_ann_learning_M
```

```r
data_ann_test_M <- data_ann_M[(test80_M + 1):n,]
head(data_ann_test_M)
tail(data_ann_test_M)
```

```r
INPUT_NODES <- 10
HIDDEN_NODES <- 5
OUTPUT_NODE <- 5
ITERATION <- 95
```

```r
getDataSet <- function(item,from,to,size){
    dataframe <- NULL
    to <- to - size + 1
    for(i in from:to){
        start <- i
        end <- start + size - 1
        temp <- item[c(start:end)]
        dataframe <- rbind(dataframe, t(temp))
    }
    return(dataframe)
}
```

```r
in_learning <- getDataSet(data_ann_learning$가격norm, 1,45, INPUT_NODES)
in_learning
```

```r
out_learning <- getDataSet(data_ann_learning$가격norm,8, 47, OUTPUT_NODE)
out_learning
```

```r
#행을 맞춰야 함
model <- nnet(in_learning, out_learning, size = HIDDEN_NODES, maxit = ITERATION)
```

```r
in_test_M <- getDataSet(data_ann_test_M$가격norm, 1,11, INPUT_NODES)
in_test_M
```

```r
predicted_values_M <- predict(model, in_test_M, type="raw")
Vpredicted_M <- (predicted_values_M - 0.05)/ 0.9 * (max(data_ann_M$평균가격) - min(data_ann_M$평균가격))
+ min(data_ann_M$평균가격)

Vpredicted_M
```

```r
Vreal_M <- getDataSet(data_ann_M$평균가격,55, 60, OUTPUT_NODE)
Vreal_M
```

```r
ERROR_M <- abs(Vreal_M - Vpredicted_M)
MAPE_M <- rowMeans(ERROR_M/Vreal_M) * 100
MAPE_M
mean(MAPE_M)
```
